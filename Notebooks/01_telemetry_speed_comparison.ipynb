```python
import os
import fastf1
from fastf1.plotting import setup_mpl
from matplotlib import pyplot as plt

# ‚úÖ Create cache directory
os.makedirs('data/cache', exist_ok=True)

# ‚úÖ Enable FastF1 cache
fastf1.Cache.enable_cache('data/cache')

# üîß Load a session (you can change the track/session later)
session = fastf1.get_session(2023, 'Monza', 'Q')  # 2023 Italian GP - Qualifying
session.load()

# üèéÔ∏è Compare two drivers
ver = session.laps.pick_driver('VER').pick_fastest()
ham = session.laps.pick_driver('HAM').pick_fastest()

# üìà Speed trace
setup_mpl()
ver_car_data = ver.get_car_data()
ham_car_data = ham.get_car_data()

plt.figure(figsize=(10, 5))

plt.plot(ver_car_data['Time'], ver_car_data['Speed'], label='Verstappen')
plt.plot(ham_car_data['Time'], ham_car_data['Speed'], label='Hamilton')

plt.xlabel('Time')
plt.ylabel('Speed in km/h')
plt.title('Speed Comparison - Monza Qualifying')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```
